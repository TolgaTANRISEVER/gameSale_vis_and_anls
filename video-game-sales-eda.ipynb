```python
def hisplot(variable):
    fig = px.histogram(videoGame,
                       x=variable,
                       color=variable,
                       title=variable,
                       width=700,
                       height=500,
                       barmode='overlay'
                       )
    fig.update_layout(bargap=0.1,barmode='stack', xaxis={'categoryorder': 'total descending'})
   
    fig.show()
    
```

```python
category1=['Platform', 'Genre' ]
for i in category1:
    hisplot(i)
```

```python

year_max_df = videoGame.groupby(['Year', 'Genre']).size().reset_index(name='count')
year_max_idx = year_max_df.groupby(['Year'])['count'].transform(max) == year_max_df['count']
year_max_genre = year_max_df[year_max_idx].reset_index(drop=True)
year_max_genre = year_max_genre.drop_duplicates(subset=["Year", "count"], keep='last').reset_index(drop=True)
genre = year_max_genre['Genre'].values
```

```python
year_sale_dx = videoGame.groupby(by=['Year', 'Genre'])['Global_Sales'].sum().reset_index()
year_sale = year_sale_dx.groupby(by=['Year'])['Global_Sales'].transform(max) == year_sale_dx['Global_Sales']
year_sale_max = year_sale_dx[year_sale].reset_index(drop=True)
# year_sale_max
genre = year_sale_max['Genre']
```

```python
fig=px.bar(year_max_genre,x="Year",y="count",barmode="overlay",color="Year",hover_name="Genre", text="Genre")
fig.show()
```

```python
comp_platform = videoGame[['Platform', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
comp_platform
```

```python
comp_platform
```

```python
# comp_table = comp_map.reset_index()
comp_table = pd.melt(comp_platform, id_vars=['Platform'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_table.head()
```

```python

fig = px.histogram(comp_table, x="Sale_Area",y="Sale_Price", animation_frame="Platform", animation_group="Platform",
            color="Sale_Area", hover_name="Platform")
fig.update_layout(bargap=0.1,barmode='stack')

fig.show()
```

```python
top_publisher = videoGame.groupby(by=['Publisher'])['Year'].count().sort_values(ascending=False).head(20)
top_publisher = pd.DataFrame(top_publisher).reset_index()
top_publisher["count"]=top_publisher["Year"]
top_publisher
```

```python
fig=px.bar(top_publisher,x="Publisher",y="count",color="Publisher")
fig.show()
```

```python
sale_plh=videoGame[["Publisher","Global_Sales"]]
sale_plh = sale_plh.groupby('Publisher')["Global_Sales"].sum().sort_values(ascending=False).head(20)
sale_plh = pd.DataFrame(sale_plh).reset_index()
sale_plh
```

```python
fig=px.bar(sale_plh,x="Publisher",y="Global_Sales",color="Publisher")
fig.show()
```

```python
comp_publisher = videoGame[['Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']]
comp_publisher.head()
```

```python
comp_publisher_tabl=comp_publisher.groupby(by=["Publisher"]).sum().reset_index().sort_values(by=['Global_Sales'], ascending=False)
comp_publisher_tabl = comp_publisher_tabl.head(20)
comp_publisher_tabl
```

```python
comp_publisher_tabl = pd.melt(comp_publisher_tabl, id_vars=['Publisher'], value_vars=['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales'], var_name='Sale_Area', value_name='Sale_Price')
comp_publisher_tabl
```

```python

fig = px.histogram(comp_publisher_tabl, x="Publisher",y="Sale_Price", animation_frame="Sale_Area", animation_group="Sale_Area",
            color="Publisher", hover_name="Publisher")
fig.update_layout(bargap=0.1,barmode='stack')

fig.show()
```

```python
top_publisher =  videoGame[['Year', 'Publisher']]
top_publisher_df = top_publisher.groupby(by=['Year', 'Publisher']).size().reset_index(name='Count')
top_publisher_idx =  top_publisher_df.groupby(by=['Year'])['Count'].transform(max) == top_publisher_df['Count']
top_publisher_count = top_publisher_df[top_publisher_idx].reset_index(drop=True)
top_publisher_count  = top_publisher_count.drop_duplicates(subset=["Year", "Count"], keep='last').reset_index(drop=True)
```

```python
top_publisher_count
```

```python
fig=px.bar(top_publisher_count,x="Year",y="Count",barmode="overlay",color="Publisher",hover_name="Publisher", text="Publisher")
fig.update_layout(bargap=0.1,barmode='stack', xaxis={'categoryorder': 'total descending'})
fig.show()

```

```python
top_sale_reg = videoGame[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
# pd.DataFrame(top_sale_reg.sum(), columns=['a', 'b'])
top_sale_reg = top_sale_reg.sum().reset_index()
top_sale_reg = top_sale_reg.rename(columns={"index": "region", 0: "sale"})
top_sale_reg
```

```python
fig=px.histogram(top_sale_reg,x="region",y="sale",color="region")
fig.update_layout(bargap=0.1,barmode='stack', xaxis={'categoryorder': 'total descending'})
fig.show()
```

```python
fig = px.pie(top_sale_reg, values='sale', names='region', title='Total revenue by region',
            hover_data=['sale'], labels={'sale':'sale'})
fig.update_layout(
    title={
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font_color':'rgba(0, 153, 0,0.5)'})
    
fig.show()
```
